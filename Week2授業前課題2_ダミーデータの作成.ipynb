# 【問題1】乱数の作成

```python
import numpy as np

mu_0 = [-3, 0] # 平均
sigma_0 = [[1.0, 0.8], [0.8, 1.0]] # 共分散行列

# 2次元正規乱数を500個生成
np.random.seed(0)
values_0 = np.random.multivariate_normal(mu_0, sigma_0, 500)
```

# 【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt

x1_0 = values_0[:,0]
x2_0 = values_0[:,1]

# 散布図を描画
%matplotlib inline
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x1_0, x2_0)
```

# 【問題3】ヒストグラムによる可視化

```python
# x1のヒストグラムを出力
plt.xlim(-7, 4)
plt.ylim(0, 130)
plt.hist(x1_0)
```

```python
# x2のヒストグラムを出力
plt.xlim(-7, 4)
plt.ylim(0, 130)
plt.hist(x2_0)
```

# 【問題4】データの追加

```python
mu_1 = [0, -3] # 平均
sigma_1 = [[1.0, 0.8], [0.8, 1.0]] # 共分散行列

# 2次元正規乱数を500個生成
np.random.seed(0)
values_1 = np.random.multivariate_normal(mu_1, sigma_1, 500)
```

```python
x1_0 = values_0[:,0]
x2_0 = values_0[:,1]

x1_1 = values_1[:,0]
x2_1 = values_1[:,1]


# 散布図を描画
%matplotlib inline
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x1_0, x2_0, label="0")
plt.scatter(x1_1, x2_1, label="1")
plt.legend(loc="upper right", fontsize=12) # 凡例表示
plt.show()
```

# 【問題5】データの結合

```python
# 複数の配列を結合する
values = np.concatenate([values_0, values_1])
print(values.shape)
```

# 【問題6】ラベル付け

```python
# ラベル付け
values_0 = np.insert(values_0, 2, np.full(500, 0), axis=1)
values_1 = np.insert(values_1, 2, np.full(500, 1), axis=1)

# 複数の配列を結合する
values = np.concatenate([values_0, values_1])
print(values.shape)
```
